```python
beta_sigmoid = np.copy(beta_i[:,0])

rng = np.random.default_rng(seed=12)

# Toronto
mu = 0

sigma = 0.1

I_synthetic[0,0] = I[0,0]

###########################################
 

FoI = np.zeros((len(tmoh),1))
    
for kk in range(1,len(tmoh)):

  FoI[kk,0] = beta_e[kk-1,0] * (E[kk-1,0] + I[kk-1,0]) / N[kk-1,0]

  S[kk,0] = S[kk-1,0] + dt*(- FoI[kk,0] * S[kk-1,0])
  E[kk,0] = E[kk-1,0] + dt*(FoI[kk,0]*S[kk-1,0] - (gamma_i + gamma_e)*E[kk-1,0])
  
  I[kk,0] = I[kk-1,0] + dt*(gamma_i*E[kk-1,0] - (gamma_r + gamma_d)*I[kk-1,0])
  R[kk,0] = R[kk-1,0] + dt*(gamma_e*E[kk-1,0] + gamma_r*I[kk-1,0])
  D[kk,0] = D[kk-1,0] + dt*(gamma_d*I[kk-1,0])
  N[kk,0] = S[kk,0] +  E[kk,0] + I[kk,0] + R[kk,0]

  if( kk%ndiv == 0):
      idxmoh = int(kk/ndiv)
      e = rng.normal(0, sigma, 1)
      I_synthetic[idxmoh,0] = I[kk,0] * ( 1 + e)



#### CHANGE HERE
### Save the synthetic data in CSV file for future use
# np.savetxt('./toronto_synthetic_data_brandon.csv', I_synthetic, delimiter=",")

# ### Save true model output
# np.savetxt('./toronto_synthetic_true_mult.csv', I, delimiter=",")
# np.savetxt('./toronto_synthetic_true_beta_mult.csv', beta_i, delimiter=",")

fig1, ax1 = plt.subplots(figsize=(6, 4))
plt.plot(t[0:160], I_synthetic[0:160,0], '*', markersize=6, label='Synthetic data')
plt.plot(tmoh, I, '-',linewidth=2, label='Model output')
# plt.plot(Data[0:200,0], 'o', markersize=6, label='Real data')
plt.legend(loc='best')
plt.xlabel('Time (days)')
plt.xlim([0,tlim])
plt.ylabel('Population')
plt.grid()
plt.savefig('./synthetic_data.pdf',bbox_inches='tight')
plt.show()



fig2, ax2 = plt.subplots(figsize=(6, 4))
ax2.set_xlim([0,tlim])
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Infection rate') 
ax2.grid()
plt.plot(tmoh, beta_i[:,0], label='Synthetic model of infection rate')
ax1.legend(loc='best')
plt.savefig('./beta_true.pdf',  bbox_inches='tight')
plt.show()
```
